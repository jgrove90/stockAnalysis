```python
import pandas as pd
import yfinance as yf
import datetime as dt
```

```python
def download_ticker_data(ticker):

    data = yf.download(ticker, period='max')

    # create features before saving
    (
    data
    .reset_index()
    .assign(Range=lambda _df: _df.High - _df.Low,
            prevclose=lambda _df: _df.Close.shift(1),
            PctChange=lambda _df: ((_df.Close - _df.prevclose) / _df.prevclose) * 100
            )
    .round(2)
    .drop(columns=['Adj Close', 'prevclose'])
    .to_parquet(f'{ticker}.parquet', engine='fastparquet')
    )
    
    return print(f'{ticker} data downloaded.')


# download_ticker_data('SPY')
```
